```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Classified Data',index_col=0)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler= StandardScaler()
```

```python
Scaler.fit(df.drop('TARGET CLASS',axis=1))
```

```python
dft=pd.DataFrame(Scaler.transform(df.drop('TARGET CLASS',axis=1)),columns= df.columns[:-1],)
```

```python
dft
```

```python
#data is preprocessed and transformed, thus ready for KNN model
```

```python
from sklearn.model_selection import train_test_split
```

```python
X= dft
y= df['TARGET CLASS']
    
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn= KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
predict= knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(predict,y_test))
```

```python
print(confusion_matrix(predict,y_test))
```

```python
#elbow method to choose correct k-value
```

```python
error_rate=[]

for i in range(1,30):
    knn= KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i= knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
error_rate
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,30),error_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.title(" K vs Error Rate")
plt.xlabel("k")
plt.ylabel("error rate")
```

```python
knn= KNeighborsClassifier(n_neighbors=17)  # we choose k=17 as it gives minimum error
```

```python
knn.fit(X_train,y_train)
pred=knn.predict(X_test)
```

```python
print(classification_report(pred,y_test))
print(confusion_matrix(pred,y_test))
```
